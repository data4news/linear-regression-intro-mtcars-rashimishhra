# Load data



```python
%%R -o mtcars

library(tidyverse)

mtcars
```

```python
%%R 

# pivot mtcars from wide to long using pivot_longer on all columns except the index
mtcars_long <- mtcars %>%
    rownames_to_column(var = "model") %>% # convert rownames (index) to a regular column %>% 
    pivot_longer(cols = -c(model, mpg), names_to = "variable", values_to = "value") 

mtcars_long
```

```python
%%R -w 1000

ggplot(mtcars_long) +
        aes(x=value, y=mpg) +
        geom_point() + 
        geom_smooth(method = "lm", se=F) +
        facet_wrap(~variable, scales = "free") 
```

# Regression model

## Linear model of `mpg~wt`

```python
%%R 

# Fit a linear model
model <- lm(mpg ~ wt, data = mtcars) 

# show the summary
print(summary(model))

# Plot the model
ggplot(mtcars, aes(x=wt, y=mpg)) +
        geom_point() + 
        geom_smooth(method = "lm", se=F) + 
        labs(
                title = "Linear model of mpg ~ wt",
                subtitle = "How much of the variance in mpg can be explained by wt?",
                x = "Weight (1000 lbs)",
                y = "Miles per gallon"
        )
```

* 👉 interpret what the p-value means
* 👉 interpret what the R-squared value means
* 👉 draw a conclusion in words

```python
p value- since the value is minimal compared to wt value, it is more in alignment with the larger data set
r squared- since the values are less than 1, the correlation seems to be strong, the share of Y variable is well-explained by X
The sample shows that there can be a high inverse correlation between the weight of mtcars and miles per gallon
```

Let's look at the residuals of this model now.

```python
%%R 

mtcars %>% 
    select(mpg, wt) %>% 
    mutate(
       predicted = predict(model), # get predictions
       residuals = resid(model)    # get residuals
    ) %>% 
    arrange(residuals) # sort by residuals
```

Which cars are least well predicted by this model?

👉 Toyota Corolla and Fiat 128 - higher residual values
